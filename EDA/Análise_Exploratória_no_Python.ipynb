## 1. Gráficos Simples com Pyplot

```python
#importar bibliotecas
import matplotlib.pyplot as plt
```

```python
#pedir ajuda sobre a função
plt.plot?
```

```python
#Gráfico de linhas simples e list comprehension


dados_pares = [num for num in range(12) if num % 2 == 0]
dados_impares = [num for num in range(12) if num % 2 != 0]


plt.plot(dados_pares, dados_impares,'-', color = '#daa520') # plt.plot(eixo x, eixo y, formato do marcador, cor do marcador)
plt.title('Gráfico de linhas simples')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.axis([-1, 15, -1, 15])
plt.show()
```

```python
# junção de dois gráficos
plt.plot(dados_pares, dados_impares,'-', color = '#daa520')
plt.scatter([12, 11, 14, 15, 18], [18, 15, 14, 11, 12], color = 'blue')
```

```python
# junção de três gráficos
plt.plot(dados_pares, dados_impares,'-', color = '#daa520', linewidth = 5)
plt.scatter([6, 5, 7, 8, 9], [9, 8, 7, 5, 6], color = 'blue')
plt.bar([1.2, 2.1, 3.4, 4.5, 5.8], [5.8, 4.5, 3.4, 4.1, 2.2], color = 'blue')
```

##2. Figure, axis e subplots

```python
#definir tamanho da figure
plt.figure(figsize = (10, 6)) #figsize = (largura, altura)
plt.plot(dados_pares, dados_impares)
```

```python
#definir cor do fundo da figure
plt.figure(figsize = (5, 3), facecolor = 'r')
plt.plot(dados_pares, dados_impares)
```

```python
#criando plot usando objetos e o figure (add_axes)
fig = plt.figure(figsize = (8,6))
ax = fig.add_axes([0.1, .1, 1, 0.4])

# gerando amostras
import numpy as np
mu, sigma = 60, 10
x = mu + sigma*np.random.randn(500)

#criando histograma
ax.hist(x, bins = 50, facecolor = 'green', alpha = 0.8 )
ax.set_xlabel("score")
ax.set_ylabel("n")
ax.set_title("histograma", fontsize = 10)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.grid(True)
```

```python
#alterando alpha para alterar transparência

#criando plot usando objetos e o figure (add_axes)
fig2 = plt.figure(figsize = (8,6))
ax2 = fig2.add_axes([0.1, .1, 1, 0.4])

# gerando amostras
import numpy as np
mu, sigma = 60, 10
x = mu + sigma*np.random.randn(500)

#criando histograma
ax2.hist(x[x>50], bins = 50, facecolor = 'green', alpha = 0.99 )
ax2.hist(x[x<=50], bins = 50, facecolor = 'green', alpha = 0.35 );

```

```python
#plotando 4 subplots no mesmo output

# Criando dados aleatórios
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Criando subplots
fig, axs = plt.subplots(2, 2)  # 2 linhas, 2 colunas

# Plotando dados em cada subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Seno')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosseno')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangente')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exponencial')

# Ajustando o layout para evitar sobreposição
plt.tight_layout()

plt.show()
```

## 3. Outros tipos de gráficos

```python
#Gráfico de barras

# criando figura e subplot
fig, axs = plt.subplots(2,2,figsize=(14,10))

#simples
axs[0, 0].bar([1, 2, 3, 4],[4, 9, 2, 11]) #sempre precisa dos 2 parámetros (posição, altura)

## com mais modificações

#alterando a cor e grossura da barra
axs[0, 1].bar([1, 2, 3, 4],[4, 9, 2, 11],
              color = 'r',
              width = 0.3)

# alterando a cor o contorno da barra
axs[1, 0].bar([1, 2, 3, 4],[4, 9, 2, 11],
              color = 'g',
              width = 0.3,
              edgecolor = 'orange')

# definindo ticks e rotulos de ticks, assim como incrementando formatações no gráfico
axs[1, 1].bar([1, 2, 3, 4],[4, 9, 2, 11],
              color = 'y',
              width = 0.3,
              edgecolor = 'black')
axs[1, 1].set_xticks([1, 2, 3, 4])  # Define os ticks do eixo x
axs[1, 1].set_xticklabels(['g1', 'g2', 'g3', 'g4'])  # Define os rótulos dos ticks no eixo x
axs[1,1].set_title("Gráfico de barras", fontsize = 10)
axs[1,1].set_xlabel("Eixo X", fontsize = 10)
axs[1,1].set_ylabel("Eixo y", fontsize = 10)
axs[1,1].spines['right'].set_visible(False)
axs[1,1].spines['top'].set_visible(False)



```

```python
#Gráfico de barras horizontal (só colocar o h na frente do bar)


# criando figura e subplot
fig, axs = plt.subplots(2,2,figsize=(14,10))

#simples
axs[0, 0].barh([1, 2, 3, 4],[4, 9, 2, 11]) #sempre precisa dos 2 parámetros (posição, altura)

## com mais modificações

#alterando a cor e grossura da barra
axs[0, 1].barh([1, 2, 3, 4],[4, 9, 2, 11],
              color = 'r',
              height = 0.3)

# alterando a cor o contorno da barra
axs[1, 0].barh([1, 2, 3, 4],[4, 9, 2, 11],
              color = 'g',
              height = 0.3,
              edgecolor = 'orange')

# definindo ticks e rotulos de ticks, assim como incrementando formatações no gráfico
axs[1, 1].barh([1, 2, 3, 4],[4, 9, 2, 11],
              color = 'y',
              height = 0.3,
              edgecolor = 'black')
axs[1, 1].set_yticks([1, 2, 3, 4])  # Define os ticks do eixo x
axs[1, 1].set_yticklabels(['g1', 'g2', 'g3', 'g4'])  # Define os rótulos dos ticks no eixo x
axs[1,1].set_title("Gráfico de barras horizontais", fontsize = 10)
axs[1,1].set_xlabel("Eixo X", fontsize = 10)
axs[1,1].set_ylabel("Eixo y", fontsize = 10)
axs[1,1].spines['right'].set_visible(False)
axs[1,1].spines['top'].set_visible(False)
```

```python
#alterando a cor de cada barra e de seu contorno
graf_barra = plt.bar([1, 2, 3],[4, 9, 2], tick_label = ['r', 'g', 'b'])
graf_barra[0].set_color('r')
graf_barra[0].set_edgecolor('black')
graf_barra[1].set_color('g')
graf_barra[1].set_edgecolor('purple')
graf_barra[2].set_color('b')
graf_barra[2].set_edgecolor('brown')
```

```python
import pandas as pd

housing = pd.read_csv('drive/MyDrive/Datasets/housing.csv')
housing.head(10)
```

```python
#verificar preço médio das casas de acordo com a proximidade com o oceano

colors = ['blue', 'green', 'red', 'orange', 'purple']
housing_median = plt.bar(x=housing.groupby('ocean_proximity')['median_house_value'].mean().index,
                         height = housing.groupby('ocean_proximity')['median_house_value'].mean(), color = colors, width= 0.354345)
plt.grid(True)
```

```python
#relação quantitativa categorica por proximidade ao oceano usando seaborn
import seaborn as sns

sns.catplot(x = 'ocean_proximity', kind = 'count', data = housing);
```

```python
#grafico de dispersão com regressão linear
sns.regplot(x = 'total_rooms', y = 'median_house_value', data = housing) #funciona melhor quando existem variáveis com mais correlação
```

```python
#importar conjunto de dados iris e usar uma list comprehension para criar a coluna ID do 1 ao 150
iris = pd.read_csv('drive/MyDrive/Datasets/iris.csv')
iris['ID'] = np.arange(1, len(iris['class'])+1, 1)
```

```python
#Gráfico de dispersão utilizando o dataset iris (importado na celula acima) para verificar a relação de comprimento entre petalas e sepalas
plt.scatter(iris['sepallength'], iris['petallength'])
plt.title('Sepal x petal length', fontsize = 12);
```

```python
#atribuido uma cor para cada iris["class"]
fig, ax = plt.subplots(figsize = (6, 5))

cores = {'Iris-setosa':'Green', 'Iris-versicolor':'red', 'Iris-virginica':'blue'}
ax.scatter(iris['sepallength'], iris['petallength'], c = iris['class'].apply(lambda x:cores[x]))

ax.set_title("Sepal x petal length", fontsize = 12)
ax.set_xlabel("sepallength")
ax.set_ylabel('petallength');
```

```python
#Gráfico de distribuição de variável, mostrando a distribuição dos quartis
plt.boxplot(iris['sepallength']);
```
