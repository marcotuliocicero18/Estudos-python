# **5. Missing e Alteração/Criação de campos**

```python
#selecionar campos para a criação de um dataset menor
df_null = fifa19[['ID', 'Name', 'Overall', 'Potential','Joined']].head(10)
df_null
```

```python
#verificar valores nulos
display(df_null.isna())
display(df_null.isna().sum())
```

```python
#função para tratar campo peso (Weight)

fifa19["Weight"].fillna("0lbs", inplace = True)

def substitui_lbs(value):
  x = value.replace('lbs', '')
  return int(x)

fifa19["Peso"] = fifa19['Weight'].apply(substitui_lbs)

fifa19['Peso'].head(10)
```

```python
#função para tratar campo salario (Wage)

def converte_salario(value):
  y = value.replace('€', '')
  if 'M' in y:
    y = y.replace('M', '')
    return float(y)*1000000
  elif 'K' in y:
    y = y.replace('K','')
    return float(y)*1000

fifa19['Salario'] = fifa19['Wage'].apply(converte_salario)

fifa19.head().sort_values(by = 'Salario', ascending = False)
```

```python
# criar campo com as médias das caracteristicas de goleiro

fifa19 ['goleiro'] = (fifa19['GKReflexes'] + fifa19['GKPositioning'] + fifa19['GKKicking'] + fifa19['GKHandling'] + fifa19['GKDiving'])/5

fifa19.head()
```

```python
# Novo campo com valor condicional a outro campo

fifa19['Age_class'] = np.where(fifa19['Age']>=30, "Velho", "Novo")

fifa19.head()
```

```python
fifa19['Age_class2'] = "Novo"
fifa19.loc[fifa19['Age']>=30, 'Age_class2'] = "Velho"

fifa19.head()
```
